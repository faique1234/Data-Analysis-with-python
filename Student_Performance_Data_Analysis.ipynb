```python
#Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/sample_data/StudentPerformance.csv")
df
```

```python
# Getting more information about the data
df.head()
df.info()
df.describe()
```

```python
print (df.isnull().sum())
print (df.duplicated().sum())
```

# Exploratory Data Analysis
####1. Distribution of Performance Index
####2. Boxplot: Extracurricular Activities vs Performance Index
####3. Scatterplot: Hours Studied vs Performance Index
####4. Correlation Heatmap (numeric features)

```python
# distribution of performance index
plt.figure(figsize=(8,6))
sns.histplot(df['Performance Index'], kde=True,bins=20,color='green')
plt.title("Distribution of Performance Index")
plt.xlabel("performance Index")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
print ("Median Performace Index: ",df["Performance Index"].median())
```

```python
#  Boxplot: Extracurricular Activities vs Performance Index
plt.figure(figsize=(10,5))
sns.boxplot(x= "Extracurricular Activities",y= 'Performance Index', data =df, palette ='Set2' )
plt.title("Extracurricular Activities vs Performance Index")
plt.xlabel("Extracurricular Activities")
plt.ylabel("Performance Index")
plt.tight_layout()
plt.show()
```

```python
# Scatterplot: Hours Studied vs Performance Index
plt.figure(figsize=(8,5))
sns.scatterplot(x= 'Hours Studied',y ='Performance Index', data =df, color='purple')
plt.title("Hours studied vs Performance index")
plt.xlabel("Hours studied")
plt.ylabel('Performance index')
plt.tight_layout()
plt.show()
```

```python
# Correlation Heatmap (numeric features)
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(numeric_only=True), annot= True, cmap= 'coolwarm')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
# Grouped Statistics (Optional Insights)

print("\nAverage Performance Index by Extracurricular Activities:\n",
      df.groupby('Extracurricular Activities')['Performance Index'].mean())
```
